1) Составьте аналитически закон распределения случайной величины, найдите математическое ожидание и дисперсию.

3) Проиллюстрируйте теорему Чебышёва для одинаково распределенных слагаемых

## Решение:$\newline$ Пункт 1: Составим закон распределения случайной величины в виде ряда распределения и найдём математическое ожидание и дисперсию с помощью известных нам формул мат. ожидания и дисперсии.

Т.к X - число испытанных приборов, то результат испытаний нам не важен => число испытанных приборов будет от 1 до 6.
Вероятность того что прибор окажется надежным равна 0,6; значит 1 прибор будем испытан только в том случае когда он оказался ненадежным, т.е будет иметь вероятность 0,4. По той-же логике 2 прибора будут испытаны только в том случае когда первый прибор оказался надежным, а второй - нет, т.е 2 прибора будут испытаны с верояностью $0,6 \cdot 0,4$. И так далее до случая с пятью приборами. В случае с шестью приборами, нам не важен результат испытания последнего прибора, т.к он в любом случае будет испытан => вероятность испытания шести приборов 6 приборов будет равна $0,6^6 + 0,6^5 \cdot 0,6 = 0,6^5 \cdot (0,6 + 0,4) = 0,6^5$.

```python
n = 6
x_znach = [k for k in range(1, n + 1)]
p1 = 0.6
p_znach = [(1 - p1), p1*(1 - p1), p1**2*(1 - p1),\
           p1**3*(1 - p1), p1**4*(1 - p1), p1**5]

print('Закон распределения СВ X в виде ряда распределения', \
      '\n', 'xk \t pk')
for k in range(len(x_znach)):
    print(x_znach[k], '\t', p_znach[k])
    
#найдём мат ожидание с помошью стандартной формулы для мат.ожидания
m = 0
for k in range(n):
    m += x_znach[k] * p_znach[k]
print('мат. ожидание равно:', m)

#найдём дисперсию с помошью упрощённой формулы для дисперсии
d = 0
for k in range(n):
    d += x_znach[k]**2 * p_znach[k]
d -= (m)**2
print('дисперсия равна:', d)
```

## Пункт 2: 
проведем компьютерное моделирование эксперимента - составим выборку из натуральных чисел от 0 до 6 размером 300, найдем набор значений этой выборки и отн. частоты и визуализируем этот статистический ряд

Выборку будем составлять следующим образом: составим массив из 300 единиц(во всех 300 испытаниях хотя-бы 1 прибор будет испытан) и будем испытывать следующие приборы. Переменная add будет индексатором того что следующий прибор работает(1 - если прибор работает, 0 - если прибор не работает). По условию, прибор исправен с вероятностью 0.6, т.е переменная add может принять значение 1 с вероятностью 0,6 и 0 с вероятностью 0,4. Далее, прибавляем add к элементу массива selection и проверяем, чему равна переменная add: если она равна 0, то мы прекращаем испытания и переходим к следующему эксперименту, в противном случае продолжаем испытания до тех пор пока количество приборов не будет равно 6.

```python
size = 300
Selection = np.full(size, 1)
n1 = 1
size1 = 1
for i in range(size):
    add = np.random.binomial(n1, p1, size1)
    Selection[i] += add
    if add > 0:
        while ((add > 0) and (Selection[i] != 6)):
            add = np.random.binomial(n1, p1, size1)
            Selection[i] += add
print('Выборка объёма 300:', Selection)
#найдем набор значений этой выборки и отн. частоты
z = np.sort(list(set(Selection)))
print('набор различных значений выборки:', '\n', 'zk = ', z)
Sel = np.bincount(Selection) / size
Sel = Sel[1:]
print('относительные частоты различых элементов выборки:', '\n', Sel)

#представим эту информацию в виде таблицы и гистограммы
print('статистический ряд:', '\n', 'zk \t pk')
for k in range(len(z)):
    print(z[k], '\t', Sel[k])

plt.bar(z, Sel, width = 0.1, label = 'Фактические зн-я')
plt.bar(x_znach, p_znach, width = 0.05, label = 'Теоретические зн-я')
plt.legend()
plt.title('визуализация статистического ряда')
plt.grid(True)
```

## Пункт 3: 
Теорема Чебышева заключается в следующем: Если $X_1, X_2, X_3, ..., X_n$ попарно независимые случайные величины, причем дисперсии их равномерно ограничены (не превышают постоянного числа C), то, как бы мало ни было положительное число $epsilon$, вероятность неравенства:
$$P(|\frac{X_1 + X_2 + X_3 + ... + X_n}{n} - \frac{M[X_1] + M[X_2] + M[X_3] + ... + M[X_n]}{n}| < \epsilon) \rightarrow 1 \newline$$ при $$n \rightarrow \infty$$
$\newline$
Т.к в данном эксперименте мат ожидания всех случайных величин одинаково и равно m, то необходимо рассмотреть выражение:
$$P(|\frac{X_1 + X_2 + X_3 + ... + X_n}{n} - m| < \epsilon) \rightarrow 1 \newline$$ при $$n \rightarrow \infty$$
$\newline$
Чтобы эту теорему продемонстрировать, построим графики мат ожидания и среднего арифметического случайных величин от размера выборки. Для того чтобы построить второй график, создадим массив суммы значений случаных величин и среднего арифметического случаных величин и будем их пополнять значениями.

```python
summ = []
summ.append(Selection[0])
sr_ar = []
sr_ar.append(Selection[0])
x = np.arange(1, 301, 1)
for i in range(1, size):
    summ.append(summ[i - 1] + Selection[i])
    sr_ar.append(summ[i] / (i + 1))
plt.figure(figsize = (9, 7))
plt.plot(x, sr_ar, 'b', label = 'Ср. арифм. сл. величин')
plt.plot(x, m + 0*x, 'r', label = 'Мат. Ожидание')
plt.title('К теореме Чебышева')
plt.grid(True)
plt.legend()
```

Из графика видим, что чем больше число испытаний, тем ближе ср. арифм. сл. величин к мат. ожиданию.

## Задание 3: 
Иван и Олег решают следующую задачу «20 футбольных команд, среди которых 4 призера (А, Б, В, Г) предыдущего первенства, по жеребьевке разбиваются на 4 занумерованные подгруппы по 5 команд.
Нужно найти вероятность события С = {в первую и вторую подгруппу не попадет ни один из призеров}». 

Решать задачу друзья начали с построения множества элементарных исходов Ω.
Иван подумал, что поскольку в событии С говорится только о призерах, то при решении задачи про остальные команды можно вовсе «забыть». И предложил образовать Ω из упорядоченных наборов длины четыре, в которых первый элемент – это номер команды, в которую попал призер А, второй элемент – номер команды, в которую попал призер Б, и т.д. 

Олег засомневался, что при таком построении множества элементарных исходов можно будет использовать классическую схему. У него есть гипотеза, что включенные в Ω исходы неравновозможны. 
Ваша задача – экспериментальным путем опровергнуть или «подтвердить» гипотезу Олега. Многократно проведите компьютерную «жеребьевку» 20 команд и сопоставьте частоты, с которыми наблюдаются 
результаты, из которых Иван предложил составить множество элементарных исходов. К каким выводам вы пришли?

## Решение: 
для начала найдём мощность множества элементарных исходов которое предложил Иван: каждый из четырёх призёров может попасть в одну из 4 подгрупп => $|\omega| = 4 \cdot 4 \cdot 4 \cdot 4 = 256$ вариантов

Следовательно, каждое из элементарных исходов должно иметь вероятность, равную $\frac{1}{256} = 0.00390625$.

Теперь проделаем многократную жребьевку, найдем частоты появления.

```python
#для того чтобы сократить код, 
#напишем вспомогательную функцию поиска элемента в массиве
amount = 10000
def search(elem, sel):
    k = 0
    for s in sel:
        if elem == s:
            k += 1
    return k

#теперь создадим множество элементарных исходов
omega = []
elems = ['1', '2', '3', '4']
for elem1 in elems:
    for elem2 in elems:
        for elem3 in elems:
            for elem4 in elems:
                omega.append(elem1 + elem2 + elem3 + elem4)

#далее создадим выборку достаточно большого размера из жребьёвок 
#и посмотрим, с какими частотами встречаются элементы из множества 
#элементарных исходов 
Selection = [''.join(random.choices(elems, k=4)) for n in range(amount)]
print('статистический ряд:', '\n', 'zk \t pk \t отличие от 1/256')
for elem in omega:
    pk = search(elem, Selection) / amount
    print(elem, '\t', pk, '\t', np.abs(pk - 0.00390625))
```

## Пункт 2: Перед тем как проиллюстрировать теорему Бернулли, найдем вероятность того что можно произвести сборку прибора.
$\newline$
Общее количество выриантов выбрать 6 деталей из 10 равно $С_{10}^6$ (т.к повтор невозможен, а порядок - не важен)
При этом, нам подходят только те случаи, когда все детали годные, либо только одна негодная. $\newline$
Всего вариантов выбрать 6 годных деталей $С_{8}^6$, а вариантов выбрать 5 годных деталей и одну негодную - $С_{8}^5 \cdot 2$, $\newline$
Тогда получаем $$P_{иск.} = \frac{С_{8}^6 + С_{8}^5 \cdot 2}{С_{10}^6} = \frac{2}{3}$$

Теперь проиллюстрируем теорему Бернулли, как делали это ранее в части 2.4. Будем считать что 0 - индикатор того что сборка прибора осуществима, а 1 - индикатор того что сборка прибора не осуществима

```python
#иллюстрация теоремы Бернулли
nn = 500
xx = np.array([0, 1])
pp = np.array([1/3, 2/3])
ff = np.zeros(nn)


for n in (np.arange(nn) + 1):
    vb = np.random.choice(xx, size = n, p = pp)
    ff[n - 1] = np.sum(vb) / n
plt.figure(figsize = (9, 7))
plt.plot(np.arange(nn) + 1, ff)
plt.hlines(2/3, - 20, nn + 20, colors = 'r', lw = 3)
plt.title('последовательность относительных частот')
plt.grid(True)
```
